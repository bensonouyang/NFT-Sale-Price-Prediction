```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
import imageio
import os
from PIL import Image


#import sys
#sys.path 
#!{sys.executable} -m pip install --upgrade pip
#!{sys.executable} -m pip install -q -U keras-tuner
import keras_tuner as kt



tr = pd.read_csv('data/XYtr.csv')
tr1 =tr.copy() 
tr1['cdate'] = pd.to_datetime(tr1['cdate']).values.astype(np.float64)/8.64e+13
tr1 = tr1.fillna(0)


pred = pd.read_csv('data/pred.csv')

te = pd.read_csv('data/Xte.csv')
te1 = te.copy()
te1['cdate'] = pd.to_datetime(te1['cdate']).values.astype(np.float64)/8.64e+13
te1 = te1.fillna(0)

```

```python
zz = np.zeros([tr1.shape[0],28,28])
yy = pd.qcut(tr1['total'], 10, labels = range(10))
```

```python
found = list()
for ii in range(tr1.shape[0]):
    if ii % 100 == 0:
        print('%d / %d' % (ii, tr1.shape[0]))
    if tr1['ext'][ii] == '.png':
        id = tr1.loc[ii,'id']
        ff = tr1.loc[ii, 'id'] + tr1.loc[ii, 'ext']
        path = 'data/images/images/' + ff
        if not os.path.isfile(path):
            continue
        
        pic = imageio.imread(path)
        pic = Image.fromarray(pic).resize((28,28))
        
        try:
            pic = np.mean(pic,axis = 2)
        except:
            pic = np.array(pic)
        
        found.append(ii)
        # print(pic.shape)
        zz[ii,:,:] = pic
```

```python
zz = zz/255
```

```python
zzte = np.zeros([tr1.shape[0],28,28])
found = list()
for ii in range(te1.shape[0]):
    if ii % 100 == 0:
        print('%d / %d' % (ii, te1.shape[0]))
    if te1['ext'][ii] == '.png':
        id = te1.loc[ii,'id']
        ff = te1.loc[ii, 'id'] + te1.loc[ii, 'ext']
        path = 'data/images/images/' + ff
        if not os.path.isfile(path):
            continue
        
        pic = imageio.imread(path)
        pic = Image.fromarray(pic).resize((28,28))
        
        try:
            pic = np.mean(pic,axis = 2)
        except:
            pic = np.array(pic)
        
        found.append(ii)
        # print(pic.shape)
        zzte[ii,:,:] = pic
```

```python
zzte = zzte/255
#zzte = pd.DataFrame(np.row_stack(zzte)) # https://www.codenong.com/49540922/
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape = (28,28)),
    tf.keras.layers.Dense(64, activation = 'relu'),
    tf.keras.layers.Dense(10, activation = 'softmax')
])
model.compile(optimizer = 'adam', loss= tf.keras.losses.SparseCategoricalCrossentropy(),
              metrics = ['accuracy'])
model.fit(zz, yy, epochs = 50)
```

```python
probability_model = tf.keras.Sequential([model, tf.keras.layers.Softmax()])
pred_i = pred.copy()
pred_i['total'] = probability_model.predict(zzte)
#pred_i.to_csv('data/pred_image.csv', index = False)
#probability_model.save_weights('/Users/tanlu/Desktop/~/2021 Fall/Stat 440/project2/stat440-21-project2/tensor_text_model.csv')
```

```python
# cnn

def create_cnn(width = 28, height = 28, regress = False):
    inputShape = (height,width)
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Conv2D(filters = 32,kernel_size = 3 ,activation = 'relu',input_shape = (28,28,1),padding = "same"))
    #model.add(tf.keras.layers.Conv2D(filters = 32,kernel_size = 3 ,activation = 'relu',input_shape = (28,28),padding = "same"))
    model.add(tf.keras.layers.MaxPooling2D(pool_size = 2))
    model.add(tf.keras.layers.Conv2D(filters = 64,kernel_size = 3 ,activation = 'relu',padding = "same"))
    model.add(tf.keras.layers.MaxPooling2D(pool_size = 2))
    model.add(tf.keras.layers.Conv2D(filters = 128,kernel_size = 3 ,activation = 'relu',padding = "same"))

    model.add(tf.keras.layers.Flatten())
    #model.add(tf.keras.layers.Dense(128, activation = 'relu'))
    #model.add(tf.keras.layers.Dropout(0.5))
    #model.add(tf.keras.layers.Dense(64, activation = 'relu'))
    #model.add(tf.keras.layers.Dropout(0.5))
    model.add(tf.keras.layers.Dense(10, activation = 'softmax'))
    #if regress:
    #    model.add(tf.keras.layers.Dense(1, activation="linear"))
    return model
```

```python
def create_mlp(dim, regress = False):
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Dense(8, input_dim = dim, activation = "relu"))
    model.add(tf.keras.layers.Dense(8, activation = "relu"))
    
    if regress:
        model.add(tf.keras.layers.Dense(1,activation = "linear"))
    return model
```

```python
probability_model = tf.keras.Sequential([cnn, tf.keras.layers.Softmax()])
```

```python
XYtr = pd.read_csv('data/XYtr.csv')


XYtr1 = XYtr.copy() # from stackoverflow
XYtr1['cdate'] = pd.to_datetime(XYtr1['cdate']).values.astype(np.float64)/8.64e+13

Y = XYtr1['total'].copy()
Xtr1 = XYtr1[['X.sales', 'cdate', 'fee1', 'fee2']].copy()

Xtr1 = Xtr1.astype(np.float64)

Xtr1 = Xtr1.fillna(0)

pred = pd.read_csv('data/pred.csv')

Xte = pd.read_csv('data/Xte.csv')

Xte1 = Xte.copy()
Xte1['cdate'] = pd.to_datetime(Xte1['cdate']).values.astype(np.float64)/8.64e+13
Xte1 = Xte1[['X.sales', 'cdate', 'fee1', 'fee2']].copy()
Xte1 = Xte1.astype(np.float64)
Xte1 = Xte1.fillna(0)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

Xtr6 = scaler.fit_transform(Xtr1)

Xte6 = scaler.transform(Xte1)


```

```python
mlp = create_mlp(Xtr6.shape[1], regress = False)
combinedInput = tf.keras.layers.concatenate([mlp.output,probability_model.output])
# combinedInput = tf.keras.layers.concatenate([mlp.output,cnn.output])

x = tf.keras.layers.Dense(4, activation="relu")(combinedInput)
x = tf.keras.layers.Dense(1, activation="linear")(x)
# our final model will accept categorical/numerical data on the MLP
# input and images on the CNN input, outputting a single value (the
# predicted price of the house)
model = tf.keras.Model(inputs=[mlp.input, probability_model.input], outputs=x)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(tsvdtr1,Y, test_size = 0.2)
X_train1, X_test1, y_train1, y_test1 = train_test_split(tsvdtr,Y, test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler1 = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_train1 = scaler1.fit_transform(X_train1)
X_test = scaler.transform(X_test)
X_test1 = scaler1.transform(X_test1)
```

```python
sgd1 = SGDRegressor(loss = 'epsilon_insensitive', alpha = 0,  epsilon = 0,max_iter = 2000).fit(X_train,y_train)
sgd2 = SGDRegressor(loss = 'epsilon_insensitive', alpha = 0,  epsilon = 0,max_iter = 2000).fit(X_train1,y_train1)
```

```python
print('%s, mae = %.4f, model score = %.4f' % (type(sgd1),mean_absolute_error(y_test,sgd1.predict(X_test)),sgd1.score(X_test,y_test)))
print('%s, mae = %.4f, model score = %.4f' % (type(sgd2),mean_absolute_error(y_test1,sgd2.predict(X_test1)),sgd2.score(X_test1,y_test1)))
```

```python
a = np.zeros(shape = (2,200))
for i in range(200):
    X_train, X_test, y_train, y_test = train_test_split(tsvdtr1,Y, test_size = 0.2)
    X_train1, X_test1, y_train1, y_test1 = train_test_split(tsvdtr,Y, test_size = 0.2)
    scaler = StandardScaler()
    scaler1 = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_train1 = scaler1.fit_transform(X_train1)
    X_test = scaler.transform(X_test)
    X_test1 = scaler1.transform(X_test1)
    sgd1 = SGDRegressor(loss = 'epsilon_insensitive', alpha = 0,  epsilon = 0,max_iter = 2000).fit(X_train,y_train)
    sgd2 = SGDRegressor(loss = 'epsilon_insensitive', alpha = 0,  epsilon = 0,max_iter = 2000).fit(X_train1,y_train1)
    a[0,i] = mean_absolute_error(y_test,sgd1.predict(X_test))
    a[1,i] = mean_absolute_error(y_test1,sgd2.predict(X_test1))

print('feature selection: ', np.mean(a[0,]),'full data: ' ,np.mean(a[1,]))
```
